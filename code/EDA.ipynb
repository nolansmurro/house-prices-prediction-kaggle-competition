# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
train = pd.read_csv('../data/cleaned_data/train_clean.csv')
```

```python
def dummies(df):categorical_columns = df.select_dtypes(include=['object']).columns
    
    df = pd.get_dummies(df, columns = categorical_columns, drop_first=True)
    
    return df
```

```python
train = dummies(train)
```

```python
plt.figure(figsize = (1,60))
sns.heatmap(train.corr()[['SalePrice']], cmap='coolwarm', fmt=".2f", vmin= -1, annot=True);
```
